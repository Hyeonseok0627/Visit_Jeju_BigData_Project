```python
import pandas as pd

data = pd.read_csv('카드소비패턴_수정.csv')
data
```

```python
import matplotlib

# Check the unique values in 'combined_region'
unique_regions = data['combined_region'].unique()
unique_regions
```

```python
from matplotlib import font_manager as fm, rc
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the uploaded font file
font_path = 'BMYEONSUNG_otf.otf'

# Add the font to Matplotlib's font manager
fm.fontManager.addfont(font_path)

# Set the font as the default for all text in Matplotlib
font_prop = fm.FontProperties(fname=font_path)
rc('font', family=font_prop.get_name())

# Now, we can try re-plotting the charts to see if the Korean text is displayed correctly.
# Re-plotting the charts with the new Korean font set
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10), sharex=True)
fig.suptitle('User Count Trends by Region and Month-Year', fontsize=16)

for idx, region in enumerate(unique_regions):
    ax = axes[idx // 2, idx % 2]
    region_data = data[data['combined_region'] == region]
    region_grouped = region_data.groupby('dtYearMonth')['userCount'].sum().reset_index()
    sns.lineplot(data=region_grouped, x='dtYearMonth', y='userCount', ax=ax)
    ax.set_title(region)
    ax.set_xlabel('Year-Month')
    ax.set_ylabel('User Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# 파일을 불러옵니다.
file_path = '카드소비패턴_수정2.csv'
data = pd.read_csv(file_path)

# 'marketType'이 '식음료'인 데이터 필터링
fnb_data = data[data['marketType'] == '식음료']

# combined_region별로 useCost 평균
mean_cost = fnb_data.groupby('combined_region')['useCost'].mean()

# 평균을 이용하여 그래프 그리기
fig, ax1 = plt.subplots()

# 합계 막대 그래프
mean_cost.plot(kind='bar', color='skyblue', ax=ax1, width=0.4, position=0, label='useCost 평균')

# 레이블, 제목, 범례 설정
ax1.set_ylabel('useCost 평균', color='skyblue')
ax1.set_title('combined_region에 따른 useCost 평균')
fig.legend(loc='upper right')

# 그래프 표시
plt.show()

```
