```python
# '주소' 컬럼을 공백을 기준으로 분리
# expand=True 옵션은 분리된 각 부분을 별도의 컬럼으로 만듭니다.
address_columns = data['address'].str.split(' ', expand=True)

# 결과 데이터프레임 생성
result_df = pd.concat([data, address_columns], axis=1)

# 새로운 엑셀 파일로 저장
output_file_path = './jeju_data_processed.csv'
result_df.to_csv(output_file_path, index=False)

output_file_path
```

```python
# 새로운 순서로 컬럼 재배열
new_order = ['name', '0', '1', '2', '3', '4','Key_money','area','Key_money/area','address']
df = data[new_order]

df
```

```python
# '지역별' 컬럼 생성을 위한 함수
def classify_region(row):
    city = row['1']
    area = row['2']
    if '읍' in area or '면' in area:
        return f'{city} 읍면'
    else:
        return f'{city} 도심'

# '지역별' 컬럼 생성
df['지역별'] = df.apply(classify_region, axis=1)

df

# 현재 작업 디렉토리에 CSV 파일로 저장
csv_file_name = '지역별_권리금.csv'

# 데이터 프레임을 CSV 파일로 저장
df.to_csv(csv_file_name, index=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rc

# Path to the uploaded font file
font_path = 'BMYEONSUNG_otf.otf'

# Add the font to Matplotlib's font manager
fm.fontManager.addfont(font_path)

# Set the font as the default for all text in Matplotlib
font_prop = fm.FontProperties(fname=font_path)
rc('font', family=font_prop.get_name())

# '지역별' 카테고리별로 'Key_money/area' 값의 평균을 계산
category_means = df.groupby('지역별')['Key_money/area'].mean()

# 막대 그래프 그리기
category_means.plot(kind='bar')
plt.title('지역별 평균 Key_money/area')
plt.xlabel('지역별')
plt.ylabel('평균 Key_money/area')
plt.show()

```

```python
# '지역별' 카테고리별로 'Key_money/area' 값의 평균을 계산
category_means1 = df.groupby('지역별')['Key_money/area'].mean()

# 평균이 가장 작은 그룹의 값을 기준으로 나머지 비율 계산
min_mean1 = category_means1.min()
normalized_means1 = category_means1 / min_mean1

# 원하는 순서대로 데이터 정렬 (예: 내림차순 정렬)
sorted_means1 = normalized_means1.sort_values(ascending=False)

# 막대 그래프 그리기
ax1 = sorted_means1.plot(kind='bar')
plt.title('지역별 평균 권리금 (비율)')
plt.xlabel('지역별')
plt.ylabel('정규화된 평균 권리금')

# 그래프의 막대 위에 비율 표시
for bar in ax1.patches:
    ax1.annotate(f'{bar.get_height():.2f}', 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='center',
                xytext=(0, 8), 
                textcoords='offset points')

plt.show()
```

```python
data2 = pd.read_csv('카드소비패턴_수정.csv')
data2

fnb_data = data2[data2['marketType'] == '식음료']

# 'combined_region' 카테고리별로 'useCost' 값의 평균을 계산
category_means2 = fnb_data.groupby('combined_region')['useCost'].mean()

# 평균이 가장 작은 그룹의 값을 기준으로 나머지 비율 계산
min_mean2 = category_means2.min()
normalized_means2 = category_means2 / min_mean2

# 원하는 순서대로 데이터 정렬 (예: 내림차순 정렬)
sorted_means2 = normalized_means2.sort_values(ascending=False)

# 막대 그래프 그리기
ax2 = sorted_means2.plot(kind='bar')
plt.title('지역별 평균 식음료 매출액 (비율)')
plt.xlabel('지역별')
plt.ylabel('정규화된 평균 매출액')

# 그래프의 막대 위에 비율 표시
for bar in ax2.patches:
    ax2.annotate(f'{bar.get_height():.2f}', 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='center',
                xytext=(0, 8), 
                textcoords='offset points')

plt.show()
```
