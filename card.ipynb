```python
import pandas as pd

# CSV 파일 불러오기
data = pd.read_csv('지역별_소비타입_연령_성별.csv')

# userType 열에서 '제주도민'을 필터링하여 제거
filtered_data = data[data['userType'] != '제주도민']

# 새로운 CSV 파일로 저장
filtered_data.to_csv('filtered_file.csv', index=False)
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# '맑은 고딕' 폰트 설정
fm._get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgun.ttf'  # 폰트 파일 위치
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)


# CSV 파일 불러오기
data = pd.read_csv('지역별_소비타입_연령_성별.csv')

# ageGroup을 순서대로 정렬
age_order = ['20대미만', '20대', '30대', '40대', '50대', '60대이상']
data['ageGroup'] = pd.Categorical(data['ageGroup'], categories=age_order, ordered=True)
age_group_data = data.groupby('ageGroup')['useCost'].mean().reset_index()

# 선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(age_group_data['ageGroup'], age_group_data['useCost'], marker='o', linestyle='-', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Average Use Cost')
plt.title('Average Use Cost by Age Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# '맑은 고딕' 폰트 설정
fm._get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgun.ttf'  # 폰트 파일 위치
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)


# 데이터 파일 불러오기
file_path = '지역별_소비타입_연령_성별.csv' # 파일 경로를 올바르게 설정해주세요
data = pd.read_csv(file_path)

# 시장 유형, 연령대, 성별별로 그룹화하고 평균 사용 금액 계산
market_grouped_data = data.groupby(['marketType', 'ageGroup', 'gender']).agg({'useCost': 'mean'}).reset_index()

# 시각화 설정
sns.set(style="whitegrid")

# 평균 사용 금액에 대한 시각화
plt.figure(figsize=(15, 6))
sns.barplot(x='marketType', y='useCost', hue='ageGroup', data=market_grouped_data)
plt.title('Average Use Cost by Market Type and Age Group')
plt.xlabel('Market Type')
plt.ylabel('Average Use Cost')
plt.legend(title='Age Group')
plt.xticks(rotation=45) # 시장 유형 라벨이 길 경우 회전시키기 위함
plt.show()
```
