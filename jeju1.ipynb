```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

font_path = 'C:/Users/it/Downloads/nanum-all/나눔 글꼴/나눔바른고딕/NanumFontSetup_TTF_BARUNGOTHIC/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)


# CSV 파일 로드
csv_file_path = './_지역별_소비타입_연령_성별_v3.csv'  # CSV 파일 경로를 지정하세요
df = pd.read_csv(csv_file_path)

# 'marketType'별로 'useCost'의 합계를 계산하고, 'gender'별로 그룹화
grouped_data = df.groupby(['marketType', 'gender'])['useCost'].sum().reset_index()

# 'marketType'별로 'useCost'의 비율 계산
grouped_data['total'] = grouped_data.groupby(['marketType'])['useCost'].transform('sum')
grouped_data['useCost_ratio'] = grouped_data['useCost'] / grouped_data['total']

# 그래프 그리기
plt.figure(figsize=(10, 6))
sns.barplot(x='marketType', y='useCost_ratio', hue='gender', data=grouped_data, palette='Set2')
plt.title('Market Type별 Gender에 따른 Use Cost 비율', fontproperties=font_prop)
plt.ylabel('Use Cost 비율', fontproperties=font_prop)
plt.xlabel('Market Type', fontproperties=font_prop)
plt.xticks(rotation=45, fontproperties=font_prop)
plt.yticks(fontproperties=font_prop)
plt.legend(title='Gender', prop=font_prop)  # 범례 추가
plt.tight_layout()

# 그래프 출력
plt.show()
```
